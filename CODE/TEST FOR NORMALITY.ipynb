# TEST FOR NORMALITY OF A POPULATION 

Inorder to check whether a sample is from normal population we can make use of following tests
- Graphical method
- Shapiro-Wilk test

## GRAPHICAL METHODS

The data are plotted against a theoretical normal distribution in such a way that the points should form an approximately straight line if the data are normally distributed. Deviations from this straight line indicate departures from normality.

EXAMPLE : 

Sample Data : 2748,2700,2655,2822,2511,3149,3257,3213,3220,2753

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

# Sample
sample = [2748, 2700, 2655, 2822, 2511, 3149, 3257, 3213, 3220, 2753]

plt.figure(figsize=(5,4))
stats.probplot(sample, dist="norm", plot=plt)
plt.title('Normal Probability Plot')
plt.show()
```

Conclusion

The data points in this plot show a reasonably straight line, suggesting that the distribution of the dataset is not far from normal. However, there is some deviation, particularly with the data points at the ends of the plot, which is common in real-world data

## SHAPIRO WILK TEST

The Shapiro-Wilk test is widely used for testing the normality of data because of its good power properties as compared to other tests. The test generates a statistic, W, which represents how well the data conform to a normal distribution; values of W close to 1 indicate that the data are closely related to a normal distribution. The p-value from the test helps decide whether to reject the null hypothesis of normality.

Null Hypothesis : $H_0: Sample \ is \ from \  a \  normal \ population$

Alternate Hypothesis : $H_0: Sample \ is \ not \ from \ normal \ population $

EXAMPLE: Do shapiro wilk test for above sample data  

```python
w_statistic, p_value = stats.shapiro(sample)

w_statistic, p_value
```

Since p value is 0.09 > 0.05 no evidence to reject null hypothesis. i.e, no evidence for alternate hypothesis. 

Sample evidence to reject assumption that the sample is from normal population.
